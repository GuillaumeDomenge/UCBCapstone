```python
df_temp2 = df_temp.loc[df_temp["inventor_id"]==inventors[0]].head(1)
df_temp3 = df_temp.loc[df_temp["inventor_id"]==inventors[1]].head(1)
df_temp4 = df_temp.loc[df_temp["inventor_id"]==inventors[2]].head(1)
df_temp5 = df_temp.loc[df_temp["inventor_id"]==inventors[3]].head(1)
df_temp6 = df_temp.loc[df_temp["inventor_id"]==inventors[4]].head(1)
```

```python
df_temp2 = pd.concat([df_temp2,df_temp3,df_temp4,df_temp5,df_temp6])
print(df_temp2)
```

```python
series1 = df_temp2["patent_id"]
all_patents = [x.split(",") for x in series1]
print(all_patents)
series2 = df_temp2["patent_date"]
all_dates = [x.split(",") for x in series2]
print(all_dates)
```

```python
inventor_list = [inventors[0]]*5+[inventors[1]]*5+[inventors[2]]*5+[inventors[3]]*5+[inventors[4]]*5
straight_ps = [item for sublist in all_patents for item in sublist]
straight_dts = [item for sublist in all_dates for item in sublist]
print(straight_ps)
print(straight_dts)
```

```python
n=24
print(straight_ps[n])
print(straight_dts[n])
```

```python
import itertools
#newdf = pd.DataFrame({"patent_id": straight_ps, "patent_date": straight_dts, "inventor_id":inventor_list})
#newdf = newdf.set_index("patent_id")
cc = list(itertools.combinations(straight_ps,2))
listl = [c[0] for c in cc]
listr = [c[1] for c in cc]
listil = [inventor_list[straight_ps.index(i)] for i in listl]
listir = [inventor_list[straight_ps.index(i)] for i in listr]
listdl = [straight_dts[straight_ps.index(i)] for i in listl]
listdr = [straight_dts[straight_ps.index(i)] for i in listr]
newdf = pd.DataFrame({"patent_1_id": listl,"patent_2_id": listr,"patent_1_date": listdl,"patent_2_date": listdr, "patent_1_inventor": listil,"patent_2_inventor": listir})
print(newdf)
```

```python
def calc_new_row(row):
    if row["patent_1_inventor"] == row["patent_2_inventor"]:
        return 1
    else:
        return 0
newdf["is_same"] = newdf.apply(calc_new_row, axis=1)
print(newdf.to_string())
```
